#Q.1 Demonstrate three different methods for creating identical 2D arrays in NumPy. Provide the code for each method and the final output after each methods

```python
import numpy as np

# Method 1: Using np.array
array1 = np.array([[1, 2, 3], [4, 5, 6]])
print("Method 1 Output:")
print(array1)


array2 = np.zeros((2, 3), dtype=int)
array2[0] = [1, 2, 3]
array2[1] = [4, 5, 6]
print("\nMethod 2 Output:")
print(array2)

array3 = np.full((2, 3), 0, dtype=int)
array3[0] = [1, 2, 3]
array3[1] = [4, 5, 6]
print("\nMethod 3 Output:")
print(array3)
```

# Q.3 Explain the following term
1. The difference in np.array, np.asarray and np.asanyarray
2. The difference between Deep copy and shallow copy


```python
# 1.


a = [1, 2, 3]
b = np.array(a)  # Creates a new array
print(b)

c = np.array([1, 2, 3])
d = np.asarray(c)  # Does not create a new copy
print(d)

e = np.matrix([[1, 2], [3, 4]])
f = np.asanyarray(e)  # Returns a matrix, not a base ndarray
print(f)
```

```python
# 2.

original = np.array([1, 2, 3])
shallow_copy = original.view()
shallow_copy[0] = 99
print(original)  # Output: [99  2  3] - original is affected

original = np.array([1, 2, 3])
deep_copy = original.copy()
deep_copy[0] = 99
print(original)  # Output: [1 2 3] - original is not affected
```

#Q. 17 Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


x = np.random.uniform(-10, 10, 100)
y = np.random.uniform(-10, 10, 100)


def get_quadrant(x, y):
    if x > 0 and y > 0:
        return 'Quadrant I'
    elif x < 0 and y > 0:
        return 'Quadrant II'
    elif x < 0 and y < 0:
        return 'Quadrant III'
    else:
        return 'Quadrant IV'


quadrants = [get_quadrant(xi, yi) for xi, yi in zip(x, y)]


plt.figure(figsize=(8, 6))
sns.scatterplot(x=x, y=y, hue= quadrants, style= quadrants, palette='Set1', s=100, alpha=0.7)


plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.legend(title='Quadrants')
plt.axhline(0, color='gray', linestyle='--', linewidth=1)  # Add horizontal line at y=0
plt.axvline(0, color='gray', linestyle='--', linewidth=1)  # Add vertical line at x=0


plt.show()


```

#Q.18  With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'SineWave Function'8

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np


output_notebook()


x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)


p = figure(title="Sine Wave Function", x_axis_label='X-axis', y_axis_label='Y-axis',
           width=800, height=400)


p.line(x, y, legend_label="Sine Wave", line_width=2, color="blue")


p.xgrid.grid_line_color = "gray"
p.ygrid.grid_line_color = "gray"
p.xgrid.grid_line_alpha = 0.5
p.ygrid.grid_line_alpha = 0.5


show(p)
```

#Q.19  Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'8

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
import pandas as pd
import numpy as np


output_notebook()


categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=5)


data = pd.DataFrame({'Category': categories, 'Value': values})


colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']
data['Color'] = colors


source = ColumnDataSource(data)


p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Category', y_axis_label='Value',
           width=800, height=400, toolbar_location=None)


p.vbar(x='Category', top='Value', width=0.9, color='Color', legend_field='Category', source=source)


hover = HoverTool()
hover.tooltips = [("Category", "@Category"), ("Value", "@Value")]
p.add_tools(hover)


p.xgrid.grid_line_color = None
p.ygrid.grid_line_alpha = 0.5


show(p)
```

#Q.21  Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set the title as 'Interactive Pie Chart'.

```python
import plotly.express as px
import numpy as np
import pandas as pd

# Generate random data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=len(categories))

# Create a DataFrame
data = pd.DataFrame({'Category': categories, 'Value': values})

# Create an interactive pie chart
fig = px.pie(data, values='Value', names='Category', title='Interactive Pie Chart',
             labels={'Value': 'Total Count'},
             hover_data=['Value'], hole=0.6)

# Update layout for better readability
fig.update_traces(textinfo='percent+label', showlegend= True)

# Show the plot
fig.show()
```
