```python
df['Packed Cell Volume'].unique()
```

```python
df['Packed Cell Volume']=pd.to_numeric(df['Packed Cell Volume'], errors ='coerce')
```

```python
df.dtypes
```

```python
df['White Blood Cell Count'] = pd.to_numeric(df['White Blood Cell Count'], errors ='coerce')
df['Red Blood Cell Count'] = pd.to_numeric(df['Red Blood Cell Count'], errors ='coerce')
```

```python
df.dtypes
```

```python
df['Packed Cell Volume'].unique()
```

```python
cat_cols = list(df.dtypes[df.dtypes=='object'].index)
num_cols = list(df.dtypes[df.dtypes!='object'].index)
```

```python
cat_cols
```

```python
num_cols
```

```python
#unique Values in cat colums

for col in cat_cols:
  print(f"the {col} has the unique colums{df[col].unique()}")
```

```python
df['Diabetes Mellitus'] = df['Diabetes Mellitus'].replace(to_replace = {' yes':'yes', '\tno':'no','\tyes':'yes'})
```

```python
df['Diabetes Mellitus'].unique()
```

```python
df['Coronary Artery Disease'] = df['Coronary Artery Disease'].replace(to_replace = {'\tno':'no'})
df['Classification'] = df['Classification'].replace(to_replace = {'ckd\t':'ckd'})
```

```python
df.head(1)
```

```python
df['Classification'] = df['Classification'].map({'ckd': 0, 'notckd' : 1})
```
